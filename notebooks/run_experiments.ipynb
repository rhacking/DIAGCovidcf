```python
from covidcf.evaluation.hyperopt import hyperopt
from covidcf.evaluation.evaluation import evaluate
from covidcf.config.base import ConfigSchema, HyperOptConfigSchema
```

```python
import yaml
for f in os.scandir(path):
    if f.name != 'meta.yaml' and f.name.endswith('.yaml'):
        with open(f.path, 'r') as fh:
            config = yaml.load(fh, Loader=yaml.SafeLoader)
        try:
            ConfigSchema.load(config)
        except:
            config = HyperOptConfigSchema.load(config)
            hyperopt(config, f.path)
```

```python
import yaml
human_names = []
for f in os.scandir(path):
    if f.name != 'meta.yaml' and f.name.endswith('.yaml'):
        with open(f.path, 'r') as fh:
            config = yaml.load(fh, Loader=yaml.SafeLoader)
        try:
            ConfigSchema.load(config)
        except:
            config = HyperOptConfigSchema.load(config)
            human_names.append(config.human_name)
```

```python
import joblib
import yaml
ids = []
for f in os.scandir(path):
    if f.name != 'meta.yaml' and f.name.endswith('.pkl'):
        study = joblib.load(f.path)
        path_to_file, filename = os.path.split(f.path)
        config_path = path_to_file + '/' + filename.split('__')[1].split('.')[0] + '.yaml'
        with open(config_path, 'r') as fh:
            config_dict = yaml.load(fh, Loader=yaml.SafeLoader)

        for k, v in study.best_params.items():
            keys = k.split('__')
            current = config_dict['parameters']
            for key in keys[:-1]:
                current = current[key]
            current[keys[-1]] = v
        del config_dict['n_trials']
        del config_dict['dataset']['validation']

        config = ConfigSchema.load(config_dict)
        ids.append(evaluate(config, config_dict, save_path=path_to_file+'/'+'result_'+filename, name=filename.split('.')[0]))
```

```python
import joblib
import yaml
import pandas as pd
ids = []
dfs = []
for f in os.scandir(path):
    if f.name != 'meta.yaml' and f.name.startswith('result_'):
        print(f.name)
        path_to_file, filename = os.path.split(f.path)
        config_path = path_to_file + '/' + filename.split('__')[1].split('.')[0] + '.yaml'
        with open(config_path, 'r') as fh:
            config_dict = yaml.load(fh, Loader=yaml.SafeLoader)
        data = joblib.load(f.path)
        data = pd.DataFrame({'y_test': data['y_test'], 'pred': data['preds_proba']})
        data['model'] = config_dict['human_name']
        dfs.append(data)
#         study = joblib.load(f.path)
#         path_to_file, filename = os.path.split(f.path)
#         config_path = path_to_file + '/' + filename.split('__')[1].split('.')[0] + '.yaml'
#         ids.append(filename.split('.')[0])
```

```python
df = pd.concat(dfs, ignore_index=True)
```

```python
df.y_test.cat.categories
```

```python
df['complex'] = df.y_test.cat.codes + 1j*df.pred
```

```python
import seaborn as sns
from sklearn.metrics import roc_auc_score
import numpy as np
def est(x):
    x = np.array(x)
    true = x.real
    pred = x.imag
    
    return roc_auc_score(true, pred)
sns.set()
sns.barplot(data=df, x='model', y='complex', estimator=est)
```

```python
from covidcf.reporting.statistics import compare_models, delong_roc_test, bootstrap_test, mcnemar_test
```
