```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle('../data/results/rumc_hyperopt_test_pcr_no_visual.pkl')
df
```

```python
preds = df[df.metric == 'predictions'].value.values[0]
data = {'model': ['a', 'b']}
data.update({f'pred_{i}': [pred, pred] for i, pred in enumerate(preds)})
data
df2 = pd.DataFrame(data)
df2['index'] = df2.index
df2
```

```python
df2 = pd.wide_to_long(df2, ['pred'], sep='_', i='index', j='sample')
df2
```

```python
df2['preds_complex'] = df2.pred.apply(lambda x: x[0] + (x[1] == 'Positief')*1j)
```

```python
import seaborn as sns
from sklearn.metrics import roc_auc_score
def test(x):
    preds = np.array(x).real
    true = np.array(x).imag

    return roc_auc_score(true, preds)
sns.barplot(data=df2, x='model', y='pred', n_boot=5000, estimator=test)
```
